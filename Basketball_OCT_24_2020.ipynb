```python
import numpy as np
from numpy.random import choice
```

```python
def compute_transition(s0, p):
    t_mat = [[0, .5*(1-p), .5*(1+p), 0, 0], 
             [0, .5*(1-p), .5*p, 0.5, 0], 
             [0, .5*(1-p), .5*p, 0, 0.5], 
             [0, 0, 0, 1, 0], 
             [0, 0, 0, 0, 1] 
            ]
    transitions = t_mat[s0]
    s1 = choice(range(len(transitions)), p=transitions)
    return s1
```

```python
best_value = None
error = 10000000
for p in np.linspace(0.0, 1.0, num=1000):
    AD = 0
    LJ = 0
    for n in range(5000):
        s0 = 0
        while True:
            s0 = compute_transition(s0, p)
            if s0 == 3:
                AD = AD + 1
                break
            if s0 == 4:
                LJ = LJ + 1
                break
    AD_p = AD/(AD + LJ)
    LJ_p = LJ/(AD + LJ)
    error_new = np.abs(AD_p - LJ_p)
    print("AD: %0.4f, LJ: %0.4f, P: %0.3f" % (AD_p, LJ_p, p))
    if error_new < error:
        error = error_new
        best_value = p
        
```

```python
print("Best Value: %0.4f, Error: %0.4f" % (best_value, error))
```
